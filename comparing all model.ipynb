**KMC**

```python
from sklearn.cluster import KMeans

# We'll use the numerical features again for clustering
X_cluster = data[['precipitation', 'temp_max', 'temp_min', 'wind']]

# Determine the optimal number of clusters (K) using the elbow method
inertia = []
# Let's test K from 1 to 10
k_range = range(1, 11)

for k in k_range:
  kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # Add n_init
  kmeans.fit(X_cluster)
  inertia.append(kmeans.inertia_)

# Plot the elbow method graph
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.xticks(k_range)
plt.grid(True)
plt.show()

# Based on the elbow plot, choose an appropriate number of clusters (e.g., 3, 4, or 5 seems reasonable)
# Let's choose K=4 as an example (adjust based on your elbow plot)
n_clusters = 4 # Adjust this based on the elbow plot result

# Apply K-Means clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10) # Add n_init
clusters = kmeans.fit_predict(X_cluster)

# Add the cluster labels back to the original dataframe for analysis
data_clustered = data.copy()
data_clustered['cluster'] = clusters

# Visualize the clusters (example using pairs of features)
sns.pairplot(data_clustered, vars=['precipitation', 'temp_max', 'temp_min', 'wind'], hue='cluster', palette='viridis')
plt.suptitle(f'K-Means Clustering with K={n_clusters}', y=1.02)
plt.show()

# You can also examine the characteristics of each cluster
print(data_clustered.groupby('cluster').mean())

# To relate clusters to weather types, you could see the distribution of original 'weather' labels within each cluster
print("\nDistribution of original 'weather' within each cluster:")
print(pd.crosstab(data_clustered['cluster'], data_clustered['weather']))

```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


# Check if n_clusters is greater than 1, as Silhouette Score is not defined for 1 cluster
if n_clusters > 1:
  silhouette_avg = silhouette_score(X_cluster, clusters)
  print(f"\nK-Means Clustering Metrics for K={n_clusters}:")
  print(f"Silhouette Score: {silhouette_avg:.4f}")


else:
  print("\nSilhouette Score is not defined for K=1 cluster.")

```
